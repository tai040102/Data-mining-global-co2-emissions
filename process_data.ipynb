```python
import pandas as pd
import numpy as np
```

Dữ liệu CO2 từ trang Our World in Data

```python
df_co2 = pd.read_csv("/content/drive/MyDrive/CS2207-KhaiThacDuLieuVaUngDung/CuoiKy/co2-data-master/owid-co2-data.csv")
df_co2
```

```python
# Filter dữ liệu
# - Khoảng thời gian từ năm 2000 trở về sau
# - Chỉ sử dụng các cột: Country, Year, CO2(tổng lượng phát thải CO2)
df_co2_filtered = df_co2[df_co2["year"] >= 2000][["country", "year", "co2"]]
df_co2_filtered
```

```python
# Chuyển đổi dataframe dưới dạng:
# - Mỗi row chứa dữ liệu co2 của từng quốc gia (cột) trong năm
df_pivot_co2 = df_co2_filtered.pivot(index="year", columns="country", values="co2")
df_pivot_co2
```

```python
# Thống kê dữ liệu
num_countries = len(df_pivot_co2.columns)
no_data_countries = df_pivot_co2.columns[df_pivot_co2.isna().all()].tolist()

missing_some_countries = df_pivot_co2.columns[df_pivot_co2.isna().any()].tolist()

full_data_countries = df_pivot_co2.columns[~df_pivot_co2.isna().any()].tolist()



print("Số quốc gia:" ,num_countries)
print("Số quốc gia không có dữ liệu CO2 ở tất cả các năm:", len(no_data_countries))
print("Danh sách:", no_data_countries)
print("Số quốc gia có ít nhất 1 giá trị CO₂ bị thiếu:", len(missing_some_countries))
print(*missing_some_countries, sep="\n")
print("Số quốc gia có dữ liệu CO₂ đầy đủ:", len(full_data_countries))
```
