```python
import pandas as pd
df = pd.read_csv("/content/car-sales.csv")
print("Kích thước ban đầu:", df.shape)
print(df.head())
df = df.dropna(subset=["Code"])
df = df[df["Code"].str.len() == 3]
print("Sau khi lọc theo quốc gia:", df.shape)
df.head()
```

```python
output_path = "/content/car-sales-clean.csv"
df.to_csv(output_path, index=False)

print("File đã được lưu tại:", output_path)
```
